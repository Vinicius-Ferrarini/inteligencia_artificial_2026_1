```python
from collections import Counter

classes = [d["classe"] for d in data]
total = len(data)

class_counts = Counter(classes)

priors = {c: class_counts[c] / total for c in class_counts}
print(priors)
```

```python
def conditional_prob(feature, value, classe, data, alpha=1):
    subset = [d for d in data if d["classe"] == classe]
    count_feature = sum(1 for d in subset if d[feature] == value)

    # número de valores possíveis da feature
    possible_values = set(d[feature] for d in data)
    k = len(possible_values)

    return (count_feature + alpha) / (len(subset) + alpha * k)
```

```python
paciente = {
    "idade": "infantil",
    "diagnostico": "hipermetropia",
    "astigmatismo": "nao",
    "lacrimal": "reduzida"
}
import math

def predict(paciente, data):
    results = {}

    for classe in class_counts:
        prob = priors[classe]

        for feature in paciente:
            prob *= conditional_prob(feature, paciente[feature], classe, data)

        results[classe] = prob

    # normalização
    total_prob = sum(results.values())
    normalized = {c: results[c] / total_prob for c in results}

    return normalized

probs = predict(paciente, data)
print(probs)
```

```python
predicted = max(probs, key=probs.get)
print("Classe prevista:", predicted)
```
